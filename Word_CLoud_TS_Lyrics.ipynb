```
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```
df = pd.read_csv("taylor_swift_lyrics.csv", index_col=0, encoding= 'unicode_escape')
```

```
df.head()
```

```
df.tail()
```

```
album = df.groupby("album")

album.describe().head()
```

```
album.head()
```

```
?WordCloud
```

```
# Start with one review:
text = df.lyric[0]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
text = " ".join(lines for lines in df.lyric)
```

```
text
```

```
# Create stopword list:
stopwords = set(STOPWORDS)


# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
reputation = " ".join(lines for lines in df[df["album"]=="reputation"].lyric)
```

```
# Create stopword list:
stopwords = set(STOPWORDS)


# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(reputation)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
album_1989 = " ".join(lines for lines in df[df["album"]=="1989"].lyric)
```

```
# Create stopword list:
stopwords = set(STOPWORDS)


# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="brown").generate(album_1989)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
red = " ".join(lines for lines in df[df["album"]=="Red"].lyric)
```

```
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["oh", "ooh"])


# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="red").generate(red)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
